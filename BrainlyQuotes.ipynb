```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.brainyquote.com/topics/positive-quotes'
response = requests.get(url)
print(response)
```

```python
soup = BeautifulSoup(response.content)
```

```python
soup.prettify
```

```python
without_img = soup.findAll('div', {'class' : 'grid-item qb clearfix bqQt'})
```

```python
print(type(without_img))
print(type(without_img[0]))
```

```python
first_msg = without_img[0].a.div.text
print(first_msg)

for msg in without_img:
    if msg is not None:
        try:
            print(msg.a.div.text)
        except Exception:
            pass
```

```python
for index, msg  in enumerate(without_img):
    if msg is not None:
        try:
            with open(f'motivation{index}.txt', 'w') as f:
                f.write(msg.a.div.text)
        except Exception:
            pass
    
```
